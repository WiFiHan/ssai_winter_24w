# 클래스와 상속

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def introduce(self):
        print(f"안녕하세요, 제 이름은 {self.name}이고 나이는 {self.age}살입니다.")

    def celebrate_birthday(self):
        self.age += 1
        print(f"생일 축하합니다! 이제 {self.age}살이에요.")
```

```python
p1 = Person("홍길동", 25) # 객체 생성
p2 = Person("김영희", 30)

p1.introduce() # 메서드 호출
# 출력: 안녕하세요, 제 이름은 홍길동이고 나이는 25살입니다.

p2.celebrate_birthday()
# 출력: 생일 축하합니다! 이제 31살이에요.
```

```python
#부모 클래스
class Animal:
    def __init__(self, name, sound):
        self.name = name
        self.sound = sound

    # Animal의 메서드
    def speak(self):
        print(f"{self.name}는 '{self.sound}' 소리를 냅니다.")

    def eat(self):
        print(f"{self.name}가 음식을 먹습니다.")
```

```python
class Dog(Animal):
    def __init__(self, name, sound, breed):
        # 부모 클래스의 __init__ 메서드 호출(상속)
        super().__init__(name, sound)
        # Dog 클래스만의 새로운 속성 추가
        self.breed = breed

    # 부모 메서드 'speak' 재정의(오버라이딩)
    def speak(self):
        print(f"{self.name}(품종: {self.breed})가 '왈왈' 소리를 냅니다.")

    #Dog만의 새로운 메서드 추가
    def fetch(self, item):
        print(f"{self.name}가 {item}을 물어옵니다.")
```
